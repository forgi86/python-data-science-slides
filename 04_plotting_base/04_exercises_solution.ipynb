# Matplotlib exercises

## Import matplotlib (and numpy)

* Import ``matplotlib.pyplot`` using the standard name ``plt``
* Import ``numpy`` using the standard name ``np``

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Creating and saving plots

* Plot the function $y=\sin(x)$ on 100 points in the interval $[0, 2\pi]$ and save it a file sinusoidal.png

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.title("y = sin(x)")
plt.savefig("sinusoidal.png")
```

* Plot the functions $\sin(x)$ and $\cos(x)$ on 100 points in the interval $[0, 2\pi]$. 

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure()
plt.plot(x, y1)
plt.plot(x, y2);
```

*  Reproduce the plot below using matplotlib (with the same title, legend, gridlines, and style options)

<img src="img/trigonometric.png" aligh="left"></img>

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure()
plt.plot(x, y1, 'k--', label="sin(x)")
plt.plot(x, y2, 'b-.', label="cos(x)")
plt.title("Trigonometric functions sin and cos")
plt.xlabel("x")
plt.ylabel("y")
plt.legend() 
#plt.legend(loc="lower left")
plt.grid();
#plt.savefig("trigonometric.png")
```

## Subplots

*  Reproduce the plot below using matplotlib 

<img src="img/trigonometric_subplots.png" aligh="left"></img>

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(2, 1, figsize=(10, 6))
plt.suptitle("Trigonometric functions sin and cos")
ax[0].plot(x, y1, 'k--', label="sin(x)")
ax[0].grid()
ax[0].set_ylabel("sin(x)")
ax[1].plot(x, y2, 'b-.', label="cos(x)")
ax[1].grid()
ax[1].set_xlabel("x")
ax[1].set_ylabel("cos(x)");
#plt.savefig("trigonometric_subplots.png")
```

## More advanced plots

Let us load the well-known iris dataset

```python
import pandas as pd
iris = pd.read_csv("iris.csv") # load the dataset
iris.sample(5) # 5 random samples from the dataset
```

The dataset represent iris flowers belonging to three species: setosa, versicolor, virginica. For each specimen, 4 measurements are recorded: sepal_length, sepal_width, petal_length, petal_width 

Is there any relation between the 4 measurements and the species? Is it possible to guess the species, given the value of the 4 measurements?

* Generate the scatterplot petal_width vs. petal_length. What do you observe?

```python
plt.figure(figsize=(10,6))
plt.plot(iris["petal_length"], iris["petal_width"], 'o');
plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)");
```

Petal width and petal length are positively correlated.

* Generate the scatterplot petal_width vs. petal_length, with different colors for the 3 species. What do you observe?

```python
 # hint: you may use boolean indexing to select the rows corresponding to a species
iris_set = iris[iris['species'] == 'setosa']
iris_ver = iris[iris['species'] == 'versicolor']
iris_vir = iris[iris['species'] == 'virginica']

plt.figure(figsize=(10,6))
plt.plot(iris_set["petal_length"], iris_set["petal_width"], 'ro', label="setosa");
plt.plot(iris_ver["petal_length"], iris_ver["petal_width"], 'go', label="versicolor");
plt.plot(iris_vir["petal_length"], iris_vir["petal_width"], 'bo', label="virginica");

plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")
plt.legend();
```

Specimens of setosa have lower petal width/petal length...

* Generate the histogram of the variable petal_length. 

```python
plt.figure(figsize=(10,6))
plt.hist(iris["petal_length"])
plt.xlabel("Petal length")
plt.ylabel("Count (-)")
plt.title("Petal length histogram");
```

* Generate three histograms of the variable petal_length for the three different classes. 

```python
plt.figure(figsize=(10,6))
plt.hist(iris_set["petal_length"], label="setosa", alpha=0.7);
plt.hist(iris_ver["petal_length"], label="versicolor", alpha=0.7);
plt.hist(iris_vir["petal_length"], label="virginica", alpha=0.7);
plt.xlabel("Petal length (cm)")
plt.legend();
```

* Generate three boxplots of the variable petal_length for the three different classes

```python
iris_lst = [iris_set['petal_length'], iris_ver['petal_length'], iris_vir['petal_length']]
plt.figure(figsize=(10,6))
plt.boxplot(iris_lst, labels=["setosa", "versicolor", "virginica"])
plt.ylabel("Petal length (cm)");
```
