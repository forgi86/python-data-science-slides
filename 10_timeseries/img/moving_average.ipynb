```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = [25, 10]
plt.rcParams['font.size'] = 30
```

```python
t = np.linspace(0, 2*np.pi, 1000)
df_sin = pd.DataFrame({"t": t, "y": np.sin(t) + 0.1*np.random.randn(1000)})
```

```python
df_sin["y_MA_standard"] = df_sin["y"].rolling(51).mean()
df_sin["y_MA_centered"] = df_sin["y"].rolling(51, center=True).mean()
```

```python
plt.figure()
plt.plot(df_sin["y"], label="signal")
plt.plot(df_sin["y_MA_standard"], label="MA(51) standard", linewidth=4)
plt.plot(df_sin["y_MA_centered"], label="MA(51) centered", linewidth=4)
plt.legend();
plt.savefig("y_MA.png")
```
